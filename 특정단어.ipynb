```python
from nltk.corpus import PlaintextCorpusReader
import matplotlib.pyplot as plt
import nltk
%matplotlib inline
```

```python
corpus_root = "/Users/jianny/AI_practice/GenomicsInformatics/raw/"
```

```python
file1 = PlaintextCorpusReader(corpus_root, 'gi-2017.*\\.txt', encoding='utf-8')
file2 = PlaintextCorpusReader(corpus_root, 'gi-2018.*\\.txt', encoding='utf-8')
file3 = PlaintextCorpusReader(corpus_root, 'gni-6.*\\.txt', encoding='utf-8')
file4 = PlaintextCorpusReader(corpus_root, 'gni-7.*\\.txt', encoding='utf-8')
file5 = PlaintextCorpusReader(corpus_root, 'gni-8.*\\.txt', encoding='utf-8')
file6 = PlaintextCorpusReader(corpus_root, 'gni-9.*\\.txt', encoding='utf-8')
file7 = PlaintextCorpusReader(corpus_root, 'gni-10.*\\.txt', encoding='utf-8')
file8 = PlaintextCorpusReader(corpus_root, 'gni-11.*\\.txt', encoding='utf-8')
file9 = PlaintextCorpusReader(corpus_root, 'gni-12.*\\.txt', encoding='utf-8')
file10 = PlaintextCorpusReader(corpus_root, 'gni-13.*\\.txt', encoding='utf-8')
file11 = PlaintextCorpusReader(corpus_root, 'gni-14.*\\.txt', encoding='utf-8')
```

```python
#연도별로 분류
file_2008 = [file3,file4,file5,file6] #2008년
file_2012 = file7 #2012년
file_2013 = file8
file_2014 = file9
file_2015 = file10
file_2016 = file11
file_2017 = file1
file_2018 = file2

#Vol 별로 분류
fileNameList1 = [file3,file4,file5,file6]
fileNameList2 = [file7,file8,file9,file10,file11]
fileNameList3 = [file1,file2]
```

```python
fileYearList = [file_2012,file_2013,file_2014,file_2015,file_2016,file_2017,file_2018]
```

```python
#연도별로 frequency 비교
year_2008 = [(target,2008)
             for fileName in file_2008
             for fileid in fileName.fileids()
             for w in fileName.words(fileid)
             for target in ['cancer','expression','algorithm','alignments','genome']
             if w.lower().startswith(target)]

year_2012 = [(target,2012)
               for fileid in file_2012.fileids()
               for w in file_2012.words(fileid)
               for target in ['cancer','expression','algorithm','alignments','genome']
               if w.lower().startswith(target)]

year_2013 = [(target,2013)
               for fileid in file_2013.fileids()
               for w in file_2013.words(fileid)
               for target in ['cancer','expression','algorithm','alignments','genome']
               if w.lower().startswith(target)]

year_2014 = [(target,2014)
               for fileid in file_2014.fileids()
               for w in file_2014.words(fileid)
               for target in ['cancer','expression','algorithm','alignments','genome']
               if w.lower().startswith(target)]

year_2015 = [(target,2015)
               for fileid in file_2015.fileids()
               for w in file_2015.words(fileid)
               for target in ['cancer','expression','algorithm','alignments','genome']
               if w.lower().startswith(target)]

year_2016 = [(target,2016)
               for fileid in file_2016.fileids()
               for w in file_2016.words(fileid)
               for target in ['cancer','expression','algorithm','alignments','genome']
               if w.lower().startswith(target)]

year_2017 = [(target,2017)
               for fileid in file_2017.fileids()
               for w in file_2017.words(fileid)
               for target in ['cancer','expression','algorithm','alignments','genome']
               if w.lower().startswith(target)]

year_2018 = [(target,2018)
               for fileid in file_2018.fileids()
               for w in file_2018.words(fileid)
               for target in ['cancer','expression','algorithm','alignments','genome']
               if w.lower().startswith(target)]
```

```python
year = year_2008 + year_2012 + year_2013 + year_2014 + year_2015 + year_2016 + year_2017 + year_2018
```

```python
cfd_year = nltk.ConditionalFreqDist(year)
```

```python
cfd_year.tabulate()
cfd_year.plot(marker='o')
```
